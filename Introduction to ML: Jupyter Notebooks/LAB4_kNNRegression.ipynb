```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
iris = load_iris()
auxiliary_data = iris.data[:, :2]  
target_data = iris.data[:, 1]  #sepal width as the target variable
X_train, X_test, y_train, y_test = train_test_split(auxiliary_data, target_data, test_size=0.2, random_state=42)
```

```python
def knn_regression(X_train, y_train, X_test, k, metric):
    knn = KNeighborsRegressor(n_neighbors=k, metric=metric)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return y_pred

k_values = [1, 3, 5, 7, 9]

for k in k_values:
    y_pred = knn_regression(X_train, y_train, X_test, k, 'euclidean')
    mse = mean_squared_error(y_test, y_pred)
    
    print(f"KNN Regression Results (k={k}):")
    print(f"MSE: {mse}")
```

```python
for k in k_values:
    knn_sklearn = KNeighborsRegressor(n_neighbors=k, metric='euclidean')
    knn_sklearn.fit(X_train, y_train)
    y_pred_sklearn = knn_sklearn.predict(X_test)
    mse_sklearn = mean_squared_error(y_test, y_pred_sklearn)
    
    print(f"Scikit-Learn KNN Regression Results (k={k}):")
    print(f"MSE: {mse_sklearn}")
```
